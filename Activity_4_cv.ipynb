```python
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
pil_img = Image.open(file_name).convert("RGB")
img = np.array(pil_img)
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
level0 = gray
level1 = cv2.pyrDown(level0)
level2 = cv2.pyrDown(level1)
level3 = cv2.pyrDown(level2)
edges0 = cv2.Canny(level0, 100, 200)
edges1 = cv2.Canny(level1, 100, 200)
edges2 = cv2.Canny(level2, 100, 200)
edges3 = cv2.Canny(level3, 100, 200)
titles = [
    "Level 0 - Original", "Edges Level 0",
    "Level 1 - Half Size", "Edges Level 1",
    "Level 2 - Quarter Size", "Edges Level 2",
    "Level 3 - One-Eighth Size", "Edges Level 3"
]
images = [
    level0, edges0,
    level1, edges1,
    level2, edges2,
    level3, edges3
]
plt.figure(figsize=(12,10))
for i in range(8):
    plt.subplot(4,2,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
